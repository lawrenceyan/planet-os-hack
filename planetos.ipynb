```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
```

```python
# shapefiles from Humanitarian Data Exchange
# https://data.humdata.org/dataset/iraq-admin-level-1-boundaries
iraq = gpd.read_file('shapefiles/iraq_admbnda_adm0.shp')
data = gpd.read_file('shapefiles/iraq_admbnda_adm2.shp')
fig = data.plot(figsize = (15, 25))
plt.axis('off')
plt.show()
```

```python
# solving the point in polygon problem
# https://automating-gis-processes.github.io/2016/Lesson3-point-in-polygon.html
p1 = Point(24.952242, 60.1696017)
poly = iraq['geometry'][0]
p1.within(poly)
```
